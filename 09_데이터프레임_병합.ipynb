```python
import pandas as pd
```

```python
# 붉은색 과일
red_data = { 'fruit': ['사과', '토마토', '딸기'] }
red_df = pd.DataFrame(red_data)
red_df
```

```python
# 노란색 과일
yellow_data = { 'fruit': ['레몬', '바나나'] }
yellow_df = pd.DataFrame(yellow_data)
yellow_df
```

```python
# 세로로 붙이기
# 동일 표현: pd.concat([red_df, yellow_df], axis=0, ignore_index=False)
pd.concat([red_df, yellow_df])
```

```python
# 두 데이터프레임 사이에 중복된 index가 있는지 검증할 수 있음
# pd.concat([red_df, yellow_df], verify_integrity=True)
```

```python
# 인덱스 새로 정렬 
pd.concat([red_df, yellow_df], ignore_index=True)
```

```python
# 가로로 붙이기 
pd.concat([red_df, yellow_df], axis=1)
```

```python
shop_data = {
    'shop_id': ['A01', 'A02', 'A03', 'A04'],
    'shop_name': ['강남점', '홍대점', '분당점', '서귀포점'],
    'region': ['서울', '서울', '경기', '제주']
}

shop_df = pd.DataFrame(shop_data)
shop_df.set_index('shop_id', inplace=True)
shop_df
```

```python
fruit_data = {
	'shop_id': ['A02', 'A01', 'A01', 'A03', 'A02'],
	'fruit': ['사과', '토마토', '딸기', '레몬', '바나나'],
	'quantity': [10, 4, 8, 3, 11]
}

fruit_df = pd.DataFrame(fruit_data)
fruit_df
```

```python
fruit_df
```

```python
# 전체 컬럼(합집합)
# 동일 표현: pd.concat([shop_df, fruit_df])
pd.concat([shop_df, fruit_df], join='outer')
```

```python
# 공통 컬럼(교집합): shop_id
pd.concat([shop_df, fruit_df], join='inner')
```

```python
pd.merge(shop_df, fruit_df, on='shop_id', how='outer')
```

```python
# 동일 표현: pd.merge(shop_df, fruit_df, on='shop_id')
pd.merge(shop_df, fruit_df, on='shop_id', how='inner') 
```

```python
# shop_id와 다른 컬럼명 store_id 사용하는 과일 데이터 
new_fruit_data = {
	'store_id': ['A02', 'A01', 'A01', 'A03', 'A02'],
	'fruit': ['사과', '토마토', '딸기', '레몬', '바나나'],
	'quantity': [10, 4, 8, 3, 11]
}

new_fruit_df = pd.DataFrame(new_fruit_data)
new_fruit_df
```

```python
pd.merge(shop_df, new_fruit_df, left_on='shop_id', right_on='store_id')
```

```python
# left: 왼쪽 인자 기준으로 병합. 여기서는 shop_df
pd.merge(shop_df, fruit_df, on='shop_id', how='left')
```

```python
# right: 오른쪽 인자 기준으로 병합. 여기서는 fruit_df
pd.merge(shop_df, fruit_df, on='shop_id', how='right')
```

```python
# 왼쪽의 인덱스와 오른쪽의 컬럼으로 병합할 때 
pd.merge(shop_df, fruit_df, left_index=True, right_on='shop_id')
```

```python
fruit_df.set_index('shop_id', inplace=True)
```

```python
# 양 쪽의 인덱스로 병합할 때 
pd.merge(shop_df, fruit_df, left_index=True, right_index=True)
```
